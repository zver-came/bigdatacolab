# Частина 3: Підготовка даних

Крок 1: Витягнемо місяць і день із поля Дата.

````python
SF['Month'] = SF['Date'].apply(lambda row: int(row[0:2]))
SF['Day'] = SF['Date'].apply(lambda row: int(row[3:5]))
````

Вивиедемо деякі значення з пепередньо сформованих стовпців і перевірими type, чи ці нові стовпці дійсно містять числові
значення.

````python
print(SF['Month'][0:2])
print(SF['Day'][0:2])

# data type check
print(type(SF['Month'][0]))
````

Крок 2: Видалимо змінні з дата фрейму SF.
